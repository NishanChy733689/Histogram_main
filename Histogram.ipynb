```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Histogram

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="bill_length_mm")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            binwidth=1)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            binwidth=3)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            binwidth=5)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            bins=30)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            kde=True)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            multiple="stack")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="step")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="poly")
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="step",
            stat="density",
            common_norm=False)
plt.show()
```

```python
sns.histplot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="step",
            stat="density",
            common_norm=True)
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.histplot(data=tips, x ="size",
            stat="percent",
            discrete=True)
plt.show()
```

```python
sns.histplot(data=tips,
             x ="day",
            shrink=0.8)
plt.show()
```

```python
sns.histplot(data=tips,
             x ="day", hue="sex",
            shrink=0.8, multiple="dodge")
plt.show()
```

```python
planets = sns.load_dataset("planets")
```

```python
planets
```

```python
sns.histplot(data=planets,
            x="distance")
plt.show()
```

```python
sns.histplot(data=planets,
            x="distance",
            log_scale=True)
plt.show()
```

```python
sns.histplot(data=planets,
            x="distance",
            log_scale=True,
            fill=False)
plt.show()
```

```python
sns.histplot(data=planets,
            x="distance",
            log_scale=True,
            element="step",
            fill=False)
plt.show()
```

```python
sns.histplot(penguins,
             y="body_mass_g",
            x="bill_depth_mm")
plt.show()
```

```python
sns.histplot(penguins,
             y="body_mass_g",
            x="bill_depth_mm",
            hue="species")
plt.show()
```

```python
sns.histplot(data=planets,
             y="distance",
            x="year", bins=30,
            discrete=(True, False),
            log_scale=(False, True))
plt.show()
```

```python
sns.histplot(data=planets,
             y="distance",
            x="year", bins=30,
            discrete=(True, False),
            log_scale=(False, True),
             thresh=None)
plt.show()
```

```python
sns.histplot(data=planets,
             y="distance",
            x="year", bins=30,
            discrete=(True, False),
            log_scale=(False, True),
             pthresh=0.5, pmax=0.9)
plt.show()
```

```python
sns.histplot(data=planets,
             y="distance",
            x="year", bins=30,
            discrete=(True, False),
            log_scale=(False, True),
             cbar=True,
             cbar_kws=dict(shrink=0.75))
plt.show()
```
